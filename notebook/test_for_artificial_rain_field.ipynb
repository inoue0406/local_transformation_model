```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import GPy
```

```python
kernel = GPy.kern.RBF(input_dim=2, variance=1, lengthscale=0.2)

np.random.seed(seed=123)
```

```python
# Define Grids
N_sim = 20
x_sim = np.linspace(-1, 1, N_sim)
y_sim = np.linspace(-1, 1, N_sim)
# Expand Grid
xy = np.array(np.meshgrid(x_sim,y_sim)).reshape(2,-1).T
```

```python
xy
```

```python
mu = np.zeros(N_sim*N_sim)
cov = kernel.K(xy, xy)
```

```python
mu.shape
```

```python
cov.shape
```

```python
# generate random variable
sim = np.random.multivariate_normal(mu, cov, size=5)
```

```python
sim.shape
```

```python
img = sim[0,].reshape([N_sim,N_sim])
```

```python
plt.imshow(img)
```
